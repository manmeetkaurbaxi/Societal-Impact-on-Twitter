```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import advertools as adv
from datetime import datetime
import seaborn as sns
import glob
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
# plt.rcParams['text.usetex'] = True
```

```python
# Data folder paths
HEALTH_ORGS_PATH = '../../data/with-sentiment-labels/health-organizations/'
LEADERS_PATH = '../../data/with-sentiment-labels/leaders/'
```

### Functions for analysis

```python
def hashtags(data,column):
     hashtag_summary = adv.extract_hashtags(data[column].astype('str').values.tolist())    
     plt.figure(figsize=(14, 6))
     plt.barh([x[0] for x in hashtag_summary['top_hashtags'][:10]][::-1],
         [x[1] for x in hashtag_summary['top_hashtags'][:10]][::-1])
    #  plt.title('Top Hashtags for {}'.format(data.name))
     plt.savefig(data.name+'.pdf', transparent=True, bbox_inches='tight')
     plt.savefig(data.name+'.png', transparent=True, bbox_inches='tight')
     plt.show()
     # return print("hashtag_overview:",hashtag_summary['overview'])
     return hashtag_summary['top_hashtags'][:10]
```

```python
 def mentions(data,column):
     mention_summary = adv.extract_mentions(data[column].astype('str').values.tolist())
     plt.figure(figsize=(14,6))
     plt.barh([x[0] for x in mention_summary['top_mentions'][:10]][::-1],
         [x[1] for x in mention_summary['top_mentions'][:10]][::-1])
    #  plt.title('Top Mentions for {}'.format(data.name))
     plt.savefig(data.name+'.pdf', transparent=True, bbox_inches='tight')
     plt.savefig(data.name+'.png', transparent=True, bbox_inches='tight')
     plt.show()
     return print("overview:",mention_summary['overview'])
```

### Reading data

```python
health_orgs_df = pd.concat([pd.read_csv(f, sep=',') for f in glob.glob(HEALTH_ORGS_PATH + "/*.csv")],ignore_index=True)
health_orgs_df.name = 'Health Organizations'
print(health_orgs_df.shape)
leaders_df = pd.concat([pd.read_csv(f, sep=',') for f in glob.glob(LEADERS_PATH + "/*.csv")],ignore_index=True)
leaders_df.name = 'Leaders'
print(leaders_df.shape)
```

```python
health_orgs_df['tweet_en'] = health_orgs_df['tweet_en'].str.replace('#stop|#unsubscribe|#nye|#keep|#take|#825|وزارة_الصحة_ووقاية_المجتمع_الإمارات|كوفيد19|كوفيد19,','',regex=True,case=False)
leaders_df['tweet_en'] = leaders_df['tweet_en'].str.replace('#stop|#unsubscribe|#nye|#keep|#take|#d|#cmin|#aestahora|#envivo','',regex=True,case=False)
```

## Macro Analysis

### Hashtags Usage

```python
hashtags(health_orgs_df, 'tweet_en')
```

```python
hashtags(leaders_df, 'tweet_en')
```

### Mentions usage

```python
mentions(health_orgs_df, 'tweet_en')
```

```python
mentions(leaders_df, 'tweet_en')
```
